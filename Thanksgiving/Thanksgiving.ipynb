```python
def StoI(value) :
    if pd.isnull(value) :
        return None
    first_value = value.split(' ')[0]
    refine_value = first_value.replace('+','')
    return int(refine_value)

data["int_age"] = data["Age"].apply(StoI)
data["int_age"].describe()
```

```python
def StoI2(value) :
    if pd.isnull(value) :
        return None
    first_value = value.split(' ')[0]
    if first_value == "Prefer" :
        return None
    first_value = first_value.replace('$','')
    first_value = first_value.replace(',','')
    return int(first_value)

data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(StoI2)
data["int_income"].describe()
```

I can't find something special. Just I infer that these data are distributed very evenly as watching mean and stdard deviation value.

```python
data[data["int_income"] < 150000]["How far will you travel for Thanksgiving?"].count()
low_far = data[data["int_income"] < 150000]["How far will you travel for Thanksgiving?"]
low_far_valc = low_far.value_counts()
low_far_per = low_far_valc/low_far_valc.sum() * 100
print(low_far_per)
```

```python
over_far = data[data["int_income"]>150000]["How far will you travel for Thanksgiving?"].value_counts()
over_far_per = over_far/over_far.sum()*100
print(over_far_per)

```

# Findings
People who have income over 150000 and don't plan to go out far in Thanksgiving is more than people who have income under 150000 and have same plan.
